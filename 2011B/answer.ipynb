```python
platform_num = [20,8,17,9,15,11]
data_list = [dataA,dataB,dataC,dataD,dataE,dataF]
list_bias = [21,101,183,329,387,486]

def get_strategy(data,num,bias):
    strategy = [[] for _ in range(num)]
    for i in range(data.shape[0]):
        strategy[i] = [i+bias-num]
        for j in range(data.shape[1]):
            if data.iloc[i,j] == 1:
                strategy[i].append(j+bias)
    return strategy

def get_except_lsit(strategy,bias):
    except_list = []
    for index,line in enumerate(strategy):
        for item in line:
            if time[index+bias-1][item-1] > 3:
                except_list.append((index+bias,item))
    return except_list
```

```python
text = ['A','B','C','D','E','F']
strategies = [0]*6
except_lists = [0]*6
for i in range(len(text)):
    print('{}区的方案如下'.format(text[i]))
    strategies[i] = get_strategy(data_list[i],platform_num[i],list_bias[i])
    print(strategies[i])
    print('{}区的特殊点情况如下'.format(text[i]))
    except_lists[i] = get_except_lsit(strategies[i],list_bias[i]-platform_num[i])
    print(except_lists[i])
```

```python
import seaborn as sns
import warnings
# 改style要在改font之前
warnings.filterwarnings('ignore')
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# def plot_distribution():
data_n = pd.read_excel("附件.xls",sheet_name="全市交通路口节点数据",usecols=[4])
num = data_n.iloc[:,0].values.tolist()

pie_data = [0]*6

platform_num = [20,8,17,9,15,11]
data_list = [dataA,dataB,dataC,dataD,dataE,dataF]
list_bias = [21,101,183,329,387,486]

def plot_distribution(data,ex_list,plat_num,arg,bias):
    n = plat_num
    
    list1 = np.array([len(item) for item in data])
    list2 = np.zeros(n)
    list3 = np.arange(n)

    plt.figure(figsize=(20,12),dpi=200)
    ax = plt.subplot()

    for item in ex_list:
        list2[item[0]-bias] += 1
    index = []
    for i,item in enumerate(list2):
        if item != 0:
            index.append(i)

    total_list = np.vstack([list1,list2,list3]).T
    df = pd.DataFrame(data=total_list, columns=['非异常情况','异常情况','站点']).astype(int)

    barwidth = 0.3
    x1 = np.arange(df.shape[0])
    x2 = x1 + barwidth

    ax1 = plt.bar(x1, df['非异常情况'], width=barwidth,label='非异常情况')
    ax2 = plt.bar(x2, df['异常情况'],width=barwidth, label='异常情况')

    temp_sum1 = 0
    temp_sum2 = 0
    for i in range(n):
        p1 = ax1.patches[i]
        p2 = ax2.patches[i]
        
        temp_sum1 += p1.get_height()*num[i]
        temp_sum2 += p2.get_height()*num[i]
        
        p1.set_color('#0f4c81')
        p2.set_color('#9bb7d4')
        
        
    for item in index:
        p1 = ax1.patches[item]

        p2 = ax2.patches[item]
        r1 = p1.get_height()/(p1.get_height()+p2.get_height())
        r2 = p2.get_height()/(p1.get_height()+p2.get_height())

        plt.text(p1.get_x()+p1.get_width()/2-0.1, p1.get_height()+0.05, s=np.round(r1,2),color='black', size=10)
        plt.text(p2.get_x()+p2.get_width()/2-0.1, p2.get_height()+0.05, s=np.round(r2,2),color='black', size=10)
    plt.xticks(list3,list(map(lambda x:str(x), list3+bias)))

    plt.legend(shadow=True,loc='upper center')
    plt.title('{}区的异常与非异常情况'.format(arg))
    
    ax.set_ylabel('')
    ax.grid(which='both', axis='y', zorder=0, color='black', linestyle=':', dashes=(2,7))
    for direction in ['top','right','left']:
        ax.spines[direction].set_visible(False)

    plt.show()
    return (temp_sum1,temp_sum2)

for i in range(6):
    pie_data[i] = plot_distribution(strategies[i],except_lists[i],platform_num[i],text[i],list_bias[i]-platform_num[i])
```
