```python
platform_num = [20,8,17,9,15,11]
data_list = [dataA,dataB,dataC,dataD,dataE,dataF]
list_bias = [21,101,183,329,387,486]

def get_strategy(data,num,bias):
    strategy = [[] for _ in range(num)]
    for i in range(data.shape[0]):
        strategy[i] = [i+bias-num]
        for j in range(data.shape[1]):
            if data.iloc[i,j] == 1:
                strategy[i].append(j+bias)
    return strategy

def get_except_lsit(strategy,bias):
    except_list = []
    for index,line in enumerate(strategy):
        for item in line:
            if time[index+bias-1][item-1] > 3:
                except_list.append((index+bias,item))
    return except_list
```

```python
def cal_avg_time(i,bias):
    temp_time_sum = [0]*plat_num_list[i]
    temp_time_std = [0]*plat_num_list[i]
    for idx,i in enumerate(strategies[i]):
        source = i[0] - bias
        time_list = [time[source][item-bias]*total_num[item-bias] for item in i]
        num_list = [total_num[item-bias] for item in i]
        # 算出每个站点的每次出警时间的期望
        plat_avg_time = np.sum(time_list)/np.sum(num_list)
        temp_time_sum[idx] = plat_avg_time
        temp_time_std[idx] = np.sum(time_list)#总工作量
    return (np.mean(temp_time_sum),np.std(temp_time_std))

quota2 = [cal_avg_time(i,list_bias[i]-platform_num[i])[0] for i in range(6)]
quota3 = [cal_avg_time(i,list_bias[i]-platform_num[i])[1] for i in range(6)]
print(quota2)
print(quota3)
```

使用层次分析法来确定权重矩阵,确定每个区的合理性指标

```python
class AHP:
    def __init__(self, array):
        # 记录矩阵相关信息
        self.array = array
        # 记录矩阵大小
        self.n = array.shape[0]
        # 初始化RI值，用于一致性检验
        self.RI_list = [0, 0, 0.52, 0.89, 1.12, 1.26, 1.36, 1.41, 1.46, 1.49, 1.52, 1.54, 1.56, 1.58,
                        1.59]
        # 矩阵的特征值和特征向量
        self.eig_val, self.eig_vector = np.linalg.eig(self.array)
        # 矩阵的最大特征值
        self.max_eig_val = np.max(self.eig_val)
        # 矩阵最大特征值对应的特征向量
        self.max_eig_vector = self.eig_vector[:, np.argmax(self.eig_val)].real
        # 矩阵的一致性指标CI
        self.CI_val = (self.max_eig_val - self.n) / (self.n - 1)
        # 矩阵的一致性比例CR
        self.CR_val = self.CI_val / (self.RI_list[self.n - 1])

    """
    一致性判断
    """

    def test_consist(self):
        # 打印矩阵的一致性指标CI和一致性比例CR
        print("判断矩阵的CI值为：" + str(self.CI_val))
        print("判断矩阵的CR值为：" + str(self.CR_val))
        # 进行一致性检验判断
        if self.n == 2:  # 当只有两个子因素的情况
            print("仅包含两个子因素，不存在一致性问题")
        else:
            if self.CR_val < 0.1:  # CR值小于0.1，可以通过一致性检验
                print("判断矩阵的CR值为" + str(self.CR_val) + ",通过一致性检验")
                return True
            else:  # CR值大于0.1, 一致性检验不通过
                print("判断矩阵的CR值为" + str(self.CR_val) + "未通过一致性检验")
                return False

    """
    特征值法求权重
    """

    def cal_weight__by_eigenvalue_method(self):
        # 将矩阵最大特征值对应的特征向量进行归一化处理就得到了权重
        array_weight = self.max_eig_vector / np.sum(self.max_eig_vector)
        # 返回权重向量的值
        return array_weight
```

一般来说区均出警效率,区均非异常系数,在合理性指标中占有比较重要的位置

|                        | 区警民配比 | 区均出警效率 | 区均出警非异常系数 | 区安全系数 | 负载均衡 |
| ---------------------- | :--------: | :----------: | :----------------: | :--------: | :------: |
| **区警民配比**         |     1      |     1/5      |        1/3         |    3/2     |   1/2    |
| **区均出警效率**       |     5      |      1       |         2          |     7      |    3     |
| **区均出警非异常系数** |     3      |     1/2      |         1          |     5      |   3/2    |
| **区安全系数**         |    2/3     |     1/7      |        1/5         |     1      |   1/3    |
| **负载均衡**           |     2      |     1/3      |        2/3         |     3      |    1     |

其中1：同样重要，3：稍微重要，5：明显重要，7强烈重要 9：极端重要

```python
weight_matrix = np.array([[1,1/5,1/3,3/2,1/2],
                         [5,1,2,7,3],
                         [3,1/2,1,5,3/2],
                         [2/3,1/7,1/5,1,1/3],
                         [2,1/3,2/3,3,1]])
ahp = AHP(weight_matrix)
ahp.test_consist()
weight_array = ahp.cal_weight__by_eigenvalue_method()
weight_array
```

```python
# 计算综合得分
data_quota =  total_df.values
score = data_quota @ weight_array
total_df['得分情况'] = score
total_df
```

```python
def plot_line(df):
    plt.figure(figsize=(12,8),dpi=200)
    ax = plt.subplot()
    ax.grid(color='gray', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
    x = np.arange(6)
    
    cols = ['#00cc9d','#7ccc00','#ff6d00','#3364ff','#8719ff','#ff4d75']
    
    for idx,i in enumerate(df.columns):
        if i != '得分情况':
            ax.plot(x, df[i].values.tolist(), '.-', label=i,color=cols[idx])
        
    plt.bar(x,df['得分情况'],color='#9bb7d4',label='得分情况')

    for direction in ['top','right','left']:
        ax.spines[direction].set_visible(False)
    plt.title('各区指标对比')
    plt.xticks(x,text)
    plt.legend()
    plt.show()
plot_line(total_df)
```

* D区的分布比较均衡,且出警效率比较高,且案发次数也比较少,故评分比B区好一点,但考虑到仍有很多超出3min情况,因此可以多加入节点,解决超时情况

* C和F区由于面积较大,且分配的站点过少,所以整体指标较差,可以考虑增设节点,j决各类情况

* B区的出警效率和均衡性相较于D区更低,可以考虑增加节点改善其负载均衡情况,同提升出警效率

* A区由于人口密度大,分配的站点数多,且出警效率,负载均衡,超时情况指标,安全系数都表现良好,分布合理
